## 🚀 Training — Mode 1: TDA Only

---
## 🧠 Training — Mode 2: Text Only
Transformer CLS embedding → Linear classifier

```python
!python run.py \
    --mode text_only \
    --data_path ./data/acl2017 \
    --epochs 10 \
    --batch_size 16 \
    --lr 2e-5
```

---
## 🔀 Training — Mode 3: Hybrid (TDA + Text)
Cross‑attention fusion of TDA and transformer embeddings
**این بهترین مدل سیستم است.**

```python
!python run.py \
    --mode hybrid \
    --data_path ./data/acl2017 \
    --epochs 20 \
    --batch_size 16 \
    --lr 1e-4 \
    --max_events 0
```
